```python
import os

mydir='Bat'

def extractBatMatchNames(myDir, format='.csv'):
    fileList = []
    print('Dir :' + myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        print(files)
        for name in files:
            print(name)
            if name.endswith(format):
                fullName = os.path.join(name)
#                 print(fullName)
                fullName = fullName.rstrip('csv')
#                 print(fullName)
                fileList.append(fullName)
    return fileList


batTeamNames = extractBatMatchNames(mydir)
```
